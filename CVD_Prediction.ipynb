# Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier , ExtraTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
```

```python
lab = LabelEncoder()
for i in cat:
    data[i] = lab.fit_transform(data[i])
    
data.head()
```

```python
x = data.drop('Heart_Disease', axis= 'columns')
y= data['Heart_Disease']
```

```python
smo = SMOTE(random_state = 5)
```

```python
x_bal , y_bal = smo.fit_resample(x,y)
```

```python
X_train , x_test, Y_train, y_test = train_test_split(x_bal, y_bal, test_size = 0.3, random_state = 5)
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train, Y_train)
```

```python
pred = log.predict(x_test)
```

```python
prob = log.predict_proba(x_test)
```

```python
print(confusion_matrix(y_test, pred))
print('\n')
print(classification_report(y_test, pred))
print('\n')
print(roc_auc_score(y_test, prob[:,1]))
```

# Decision Tree

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,Y_train)
```

```python
tree_pred = dtree.predict(x_test)
tree_prob = dtree.predict_proba(x_test)
```

```python
print(confusion_matrix(y_test, tree_pred))
print('\n')
print(classification_report(y_test, tree_pred))
print(roc_auc_score(y_test, tree_prob[:,1]))
```
